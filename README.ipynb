# Simple Yeshiva Website Project

This notebook explains the structure and functionality of a simple web application built using Node.js and Express, designed as a boilerplate setup for a yeshiva. This project provides basic user registration and login functionality, along with session management using cookies and SQLite for data storage.

## Features

* User registration and login
* Session management using cookies
* SQLite database for storing user credentials and sessions
* Basic error handling

## Project Structure

The project is organized as follows:

```

│
├── index.js              # Main server file containing application logic
├── package.json          # Project metadata and dependencies
├── data.sqlite           # SQLite database file (created automatically)
└── public/               # Directory for static files
    ├── home.html        # Home page HTML
    ├── login.html       # Login page HTML
    └── register.html    # Registration page HTML
```

## Installation

Follow these steps to set up the project on your local machine:

1. **Download the Repository and unzip**:
Click on the following link to download the zipped repository: [Download SimpleWebsiteYeshivaProject](https://github.com/iTSiolinux/archive/refs/heads/main.zip).
    * **Unzip the File**:
    Once the download is complete, locate the `main.zip` file in your downloads folder and then unzip it.
    * **Access the Unzipped Repository**:
    Navigate to the newly created folder (`SimpleWebsiteYeshivaProject-main`) to access the project files.
    * Your browser will start downloading a file named `main.zip`.
2. **Install Dependencies**:
Navigate to the project directory and install the required dependencies:

``` bash
npm install
```

3. **Run the Application**:
You can start the server using the following command:

``` bash
npm start
```

Alternatively, for development mode (with auto-reload), use:

``` bash
npm run dev
```

4. **Access the Application**:
Open your web browser and navigate to `http://localhost:5000`. You should see the home page or be redirected to the login page if you are not logged in.

## Usage

* **Register**: Go to `/register` to create a new user account.
* **Login**: Go to `/login` to log in with your credentials.
* **Home**: Visit `/` to view your session ID. If you're not logged in, you'll be redirected to `/login`.
* **Logout**: `/logout` Click the logout link to end your session.
